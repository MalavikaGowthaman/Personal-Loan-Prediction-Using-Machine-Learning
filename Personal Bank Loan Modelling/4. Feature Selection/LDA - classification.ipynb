```python

# import lib
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Pre_loan_data.csv')
df = dataset
indep_X =  df.drop('Personal Loan',axis =1)
dep_Y = df['Personal Loan']
```

```python
# spliting the input and output, to perform std scaler
def split_scaler(indep_X,dep_Y):
    from sklearn.model_selection import train_test_split
    from sklearn.preprocessing import StandardScaler
    X_train,X_test,Y_train,Y_test = train_test_split(indep_X,dep_Y,test_size = 0.25,random_state = 0)
    sc= StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    return X_train,X_test,Y_train,Y_test
```

```python
# to apply pca

def lda(X_train,Y_train,n):
    from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
    lda = LDA(n_components = n)
    X_train = lda.fit_transform(X_train,Y_train)
    return lda,X_train
    
# confusion matrix
def cm_pred(classifier,X_test):
    test_pred = classifier.predict(X_test)
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(Y_test,test_pred)
    from sklearn.metrics import accuracy_score
    accuracy = accuracy_score(Y_test,test_pred)
    return classifier,cm, accuracy
```

```python
# logistics regression

def log(X_train,X_test,Y_train):
    from sklearn.linear_model import LogisticRegression
    classifier = LogisticRegression(random_state = 0)
    classifier.fit(X_train, Y_train)
    classifier,cm,accuracy = cm_pred(classifier,X_test)
    return classifier,cm,accuracy
# svm classfier

def svm_linear(X_train,X_test,Y_train):
    from sklearn.svm import SVC
    classifier = SVC(kernel = 'linear',random_state =0)
    classifier.fit(X_train, Y_train)
    classifier,cm,accuracy = cm_pred(classifier,X_test)
    return classifier,cm,accuracy
# svm classfier

def svm_nonlinear(X_train,X_test,Y_train):
    from sklearn.svm import SVC
    classifier = SVC(kernel = 'rbf',random_state =0)
    classifier.fit(X_train, Y_train)
    classifier,cm,accuracy = cm_pred(classifier,X_test)
    return classifier,cm,accuracy
# Navie Bayes classfier

def navie(X_train,X_test,Y_train):
    from sklearn.naive_bayes import GaussianNB
    classifier = GaussianNB()
    classifier.fit(X_train, Y_train)
    classifier,cm,accuracy = cm_pred(classifier,X_test)
    return classifier,cm,accuracy
# knn classfier

def knn(X_train,X_test,Y_train):
    from sklearn.neighbors import KNeighborsClassifier
    classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
    classifier.fit(X_train, Y_train)
    classifier,cm,accuracy = cm_pred(classifier,X_test)
    return classifier,cm,accuracy
# random forest classfier

def random(X_train,X_test,Y_train):
    from sklearn.ensemble import RandomForestClassifier
    classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    classifier.fit(X_train, Y_train)
    classifier,cm,accuracy = cm_pred(classifier,X_test)
    return classifier,cm,accuracy
# decision tree classfier

def decision(X_train,X_test,Y_train):
    from sklearn.tree import DecisionTreeClassifier
    classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    classifier.fit(X_train, Y_train)
    classifier,cm,accuracy = cm_pred(classifier,X_test)
    return classifier,cm,accuracy
```

```python
# to create df

def lda_df(acclog,accsvml,accsvmnl,accnb,accknn,accrf,accdt):
    df2 = pd.DataFrame(index =['LDA = 9'],columns =['Logistic','SVMLinear','SVMNL','NavieBayes','KNN','RandomForest','DecisionTree'])
    for number,index in enumerate(df2.index):
        df2['Logistic'][index]=acclog[number]
        df2['SVMLinear'][index]=accsvml[number]
        df2['SVMNL'][index]=accsvmnl[number]
        df2['NavieBayes'][index]=accnb[number]
        df2['KNN'][index]=accknn[number]
        df2['RandomForest'][index]=accrf[number]
        df2['DecisionTree'][index]=accdt[number]
    return df2

lda_= lda(indep_X,dep_Y,1)
acclog=[]
accsvml =[]
accsvmnl =[]
accnb =[]
accknn =[]
accrf =[]
accdt =[]
X_train,X_test,Y_train,Y_test = split_scaler(indep_X,dep_Y)

classifier,cm,accuracy = log(X_train,X_test,Y_train)
acclog.append(accuracy)

classifier,cm,accuracy = svm_linear(X_train,X_test,Y_train)
accsvml.append(accuracy)

classifier,cm,accuracy = svm_nonlinear(X_train,X_test,Y_train)
accsvmnl.append(accuracy)

classifier,cm,accuracy = navie(X_train,X_test,Y_train)
accnb.append(accuracy)

classifier,cm,accuracy = knn(X_train,X_test,Y_train)
accknn.append(accuracy)

classifier,cm,accuracy = random(X_train,X_test,Y_train)
accrf.append(accuracy)

classifier,cm,accuracy = decision(X_train,X_test,Y_train)
accdt.append(accuracy)

```

```python
result = lda_df(acclog,accsvml,accsvmnl,accnb,accknn,accrf,accdt)
result
```
